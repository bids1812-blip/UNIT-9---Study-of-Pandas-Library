```python
#Create DataFrame
data = {
    "Name": ["A", "B", "C"],
    "Marks": [85,90,78]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Add a new column
df["Grade"] = ["First Class", "Distinction", "Second Class"]
df
```

```python
#Update Data in DataFrame
#Change marks of student  to 88
df.loc[0,"Marks"] = 88
df
```

```python
#Delete a Column
df.drop("Grade", axis=1, inplace=True)
df
```
